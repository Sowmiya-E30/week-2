```python
plt.figure(figsize=(15,30))
for i, column in enumerate(df.columns):
  plt.subplot(7,3,i+1)
  plt.scatter(df[column],df['generated_power_kw'])
  plt.title(f'{column} vs generated_power_kw')
  plt.xlabel(column)
  plt.ylabel('generated_power_kw')
plt.tight_layout()
plt.show()
```

```python
#outlier
plt.figure(figsize=(15,30))
for i,column in enumerate(df.columns):
  plt.subplot(7,3,i+1)
  sns.boxplot(df[column],color='red')
  plt.title(f'{column} boxplot')
```

```python
##IQR##
Q1 = df[column].quantile(0.25)
Q3 = df[column].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df[column] = np.where((df[column] < lower_bound) | (df[column] > upper_bound), np.nan, df[column])
print(f'{column} outliers removed')

plt.figure(figsize=(6, 4))
sns.boxplot(y=df[column], color="lightblue")
plt.title(f"{column} Box Plot (After Removing Outliers)")
plt.show()
```
